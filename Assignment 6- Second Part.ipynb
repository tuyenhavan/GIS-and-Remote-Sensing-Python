```python
import pandas as pd
import os
from datetime import datetime
```

```python
# Set the folder path
path=r"E:\Python_Tutorials\Assignment_6"
# Place the excel file path 
file="octombrie.xlsx"
# Read the excel file
df=pd.read_excel(os.path.join(path,file)) # Skip the first five rows
df.head()
```

```python
# Select working columns including Date and Time, Description # 1 and Description # 2
df=df.loc[:,["Date and time","Description #1","Description #2"]]
df.head()
# Rename columns
df.rename(columns={"Date and time":"Date_Time","Description #1":"Room","Description #2":"Employee"},inplace=True)
df.head()
```

```python
df[["Num1","Num2","Activity"]]=df.Room.str.split("-",expand=True)
```

```python
df=df.iloc[:,[0,2,5]]
df.head()
```

```python
df=df.drop_duplicates()

df=df.sort_values("Date_Time")
# Reset index 
df=df.reset_index(drop=True) # Reset index to start at 0

df1=df
df1.head()
```

```python
len(df)
```

```python
df1['Date'] = [d.date() for d in df1['Date_Time']]

df1.head()
```

```python
def working_hour(entry,exit):
    hours=round((exit-entry).seconds/3600,2)
    return hours
```

```python
df_processed=df1
```

```python
date_unique=sorted(df_processed.Date.unique())

employee_list=set(df1.Employee)

df1["Room"]=df1.Activity
df1.drop(columns=["Activity"], inplace=True)
df1.head()
```

```python
my_date=[]

for i in df1.Date:
    dt=datetime.strftime(i,"%Y-%m-%d")
    my_date.append(dt)
df1["Date"]=my_date
df1.head()
```

```python
date_unique=sorted(df_processed.Date.unique())

Final=[]

for x in df1.Employee.unique():
    if x in employee_list:
        ketqua={}
        for y in range(len(date_unique)):
            subset=df1.query("Employee=='{0}' & Date=='{1}'".format(x,date_unique[y]))
            subset.reset_index(drop=True,inplace=True)
            subset_date=subset.sort_values("Date_Time")
            for z in range(len(subset_date)):
                if z!=0:
                    if "OUT" in subset_date.Room[z]:
                        if "IN" in subset_date.Room[z-1]:
                            exit=subset_date.Room[z-1].replace("IN","OUT")
                            if subset_date.Room[z]==exit:
                                try:
                                    formula=working_hour(subset_date.Date_Time[z-1],subset_date.Date_Time[z])
                                    if date_unique[y] in ketqua:
                                        ketqua[date_unique[y]].append(formula)
                                    else:
                                        ketqua[date_unique[y]]=[formula]
                                except:
                                    pass
        for i in ketqua:
            Final.append([x,i,sum(ketqua[i])])
```

```python
work_hour=pd.DataFrame(Final,columns=["Employee","Date","Active_Hour"])
```
