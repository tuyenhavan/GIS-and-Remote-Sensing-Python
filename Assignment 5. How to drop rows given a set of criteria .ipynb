```python
# Load the data
path=r"C:\Users\DELL\OneDrive - tuaf.edu.vn\TUAF\Python_Tutorial\DAAD Scholarship\consolidated1PT.shp"

myfile = gpd.read_file(path)
myfile.head()
```

```python
# Check the unique values of ANNOTATION column

myfile.ANNOTATION.value_counts(dropna=False, sort=False)
```

```python
# Calculate the number of unique values in a pandas.series 
print("The number of ANNOTATION unique value is {0}".format(myfile.ANNOTATION.nunique(dropna=False)))
```

```python
# Obtain of a list of ANNOTATION unique values

mlist=[]

for eachlitem in myfile.ANNOTATION.unique():
    mlist.append(eachlitem)
    
len(mlist)
# Encode each unique value with a unique number, ranging from 1 to 68

mdict={}

for eachline in myfile.ANNOTATION:
    if eachline in mdict:
        pass
    else:
        mdict[eachline]=len(mdict)+1
        
```

```python
# Encode unique value and create a new column "ANNOTATION_Code" to myfile 
mcol=[]

for eachrow in myfile.ANNOTATION:
    for eachitem in mdict:
        if eachrow==eachitem:
            mcol.append(mdict["{0}".format(eachitem)])
        else:
            pass

myfile["ANNOTATION_Code"]=mcol

myfile.head(10)
```
